
#  IPL Auction Simulator

Hi 

This is a simple Python project where I created a basic IPL Auction Simulator.
I built this project to practice Python programming and understand how real-world systems work.

The goal of this project is to simulate how IPL teams buy players in an auction.

---

##  What this project does

* Generates 150 cricket players using Python
* Creates 8 IPL teams
* Each team gets a fixed budget
* Teams randomly bid for players
* Some players are sold
* Some players remain unsold
* Final results are saved in CSV files

It works like a small version of the real IPL auction.

---

##  Teams in this project

The auction includes these 8 teams:

* SRH
* CSK
* MI
* RCB
* KKR
* RR
* DC
* GT

Each team starts with a purse of **2000**.

---

##  Players

The system creates:

* 100 Indian players
* 50 Overseas players

Each player has:

* Name
* Country
* Role (Batsman, Bowler, Allrounder, Wicketkeeper)
* Base price
* Rating

All players are generated automatically using Python.

---

##  Project Structure

```
Generate_Players.py   → Generates players
player.py             → Player class
team.py               → Team class
utils.py              → Loads players from CSV
auction.py            → Auction logic
main.py               → Runs the project

data/
  players.csv
  team_squads.csv
  sold_players.csv
  unsold_players.csv
```

---

##  How to Run

Step 1: Generate players

```
python Generate_Players.py
```

Step 2: Run the auction

```
python main.py
```

Step 3: Check the results inside the **data folder**.

---

##  Output Files

After running the project, you will get:

* team_squads.csv → Players bought by each team
* sold_players.csv → List of sold players
* unsold_players.csv → Players who were not sold

You can open these files in Excel.
